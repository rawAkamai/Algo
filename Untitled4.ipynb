```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('time_data/walmart_stock.csv')
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.head()
```

```python
df.set_index(df['Date'], inplace=True)
df.head()
```

```python
df['Close'].resample(rule='M').mean().plot(kind='bar',figsize= (16,6))
```

```python
df1 = pd.read_csv('time_data/walmart_stock.csv', index_col='Date', parse_dates=True)
df1['Open'].plot(figsize=(16,8))
df1.rolling(window=8).mean()['Close'].plot()
```
